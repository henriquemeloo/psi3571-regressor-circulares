```python
import pandas as pd
```

```python
df = pd.read_csv('./raw/weather_daily_sao_paulo.txt',sep=';')
df.drop(columns=['Unnamed: 6'], inplace=True)
df.rename(columns={
    'Estacao':'station',
    'Data':'date',
    'Hora':'time',
    'Precipitacao':'precipitation',
    'TempMaxima':'max_temperature',
    'TempMinima':'min_temperature'
}, inplace=True)
df.drop(columns=['station','time'], inplace=True)
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')
df
```

```python
df = df.groupby(['date']).sum()
df
```

```python
df.to_csv('./processed/weather_daily_sao_paulo_clean.csv', sep=';', index_label='date')
```
