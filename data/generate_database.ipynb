```python
import pandas as pd
```

```python
weather_data = pd.read_csv('./processed/weather_daily_sao_paulo_clean.csv', sep=';', index_col='date')
weather_data = weather_data.iloc[181:-3]
weather_data
```

```python
sptrans_data = pd.read_csv('./processed/pass_transp_clean.csv')
sptrans_data.rename(columns={'Unnamed: 0':'date'}, inplace=True)
sptrans_data.set_index('date', inplace=True)
sptrans_data
```

```python
date_data = pd.DataFrame(columns=['weekday','week','month'], index=pd.date_range(
    start=pd.datetime(year=2013, month=7, day=1),
    end=pd.datetime(year=2019, month=4, day=27)
))
date_data['weekday'] = date_data.index.weekday
date_data['week'] = date_data.index.week
date_data['month'] = date_data.index.month

date_data = pd.concat([date_data,
                       pd.get_dummies(date_data['weekday'], prefix='weekday'),
                       pd.get_dummies(date_data['month'], prefix='month')
                      ], axis=1)
date_data.drop(columns=['weekday', 'month'], inplace=True)

date_data.sample(5)
```

```python
database = pd.concat([date_data, weather_data, sptrans_data], axis=1)
database.sample(5)
```

```python
database.isna().sum()
```

```python
database.dropna(inplace=True)
database.shape
```

```python
database.to_csv('./processed/circular_database.csv')
```
